```python
records = sc.textFile("hdfs://devenv/user/spark/spark101/avg_temperature/data")
```

```python
def is_good(record):
    try:
        temp = int(record.split(",")[10])
    except ValueError:
        return False
    return True
```

```python
good_records = records.filter(is_good)
```

```python
day_temp = good_records.map(lambda x: (x.split(",")[1], int(x.split(",")[10])))
```

```python
day_temp.collect()
```

```python
#combineByKey(createCombiner, mergeValue, mergeCombiners, numPartitions=None, partitionFunc=<function portable_hash at 0x7f1ac7340578>)
# createCombiner='20070101',(23,1)   
# mergeValue=('20070101',(23+21,1+1) (acc[0]+v,acc[1]+1)
# mergeCombiners=((23,1))
result = day_temp.combineByKey(lambda v: (v, 1),
                               lambda acc, v: (acc[0] + v, acc[1] + 1),
                               lambda acc1, acc2: (acc1[0] + acc2[0], acc1[1] + acc2[1])) \
                 .map(lambda x: (x[0], x[1][0] / x[1][1]))
```
